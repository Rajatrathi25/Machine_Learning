```python
import webbrowser
import os
import bs4
import urllib.request as req 
```

```python
name=input("Enter the Movie name")
link='https://www.imdb.com/find?ref_=nv_sr_fn&q='+name+'&s=all'
webbrowser.open(link)
```

```python
httpresponse=req.urlopen(link)
```

```python
page=bs4.BeautifulSoup(httpresponse)
```

```python
page
```

```python
#movienames=page.find('div',class_='findMoreMatches')
#newlink=movienames.find('a')['href']
movienames=page.find('table',class_='findList')
```

```python
#newresponse=req.urlopen(link+newlink)
list1=movienames.find_all('tr')
#for i in list1:
#   print (i.text)
j=1
for i in list1:
    print(str(j)+'.'+i.text)
    j+=1
choice=int(input('Enter your choice number '))
newchoice=str(choice)
```

```python
#page1=bs4.BeautifulSoup(newresponse)
link=page.find('table',class_='findList')
#print(link)
flink=link.find_all('a')[choice]['href']
#print(flink)
newlink='https://www.imdb.com'+flink
webbrowser.open(newlink)
```

```python
#page1
newresponse=req.urlopen(newlink)
```

```python
newpage=bs4.BeautifulSoup(newresponse)
```

```python
newpage
```

```python
title=newpage.find('div',class_='title_wrapper')
```

```python
title1=title.find('h1')
newtitle=title1.text
```

```python
time=newpage.find('div',class_='subtext')
time1=time.find('time')
realtime=time1.text
realtime = realtime.replace("\n","")
realtime = realtime.replace("  ","")
```

```python
rating=newpage.find('div',class_='ratingValue')
rating1=rating.find('strong')
rating=rating1.text
```

```python
print('Name of Movie is ',newtitle)
print('Duration of Movie is ',realtime)
print('Rating of Movie is ',rating)
```
